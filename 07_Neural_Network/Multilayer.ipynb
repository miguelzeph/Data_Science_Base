This function returns the dot product of two arrays. 
For 2-D vectors, it is the equivalent to matrix multiplication.
For 1-D arrays, it is the inner product of the vectors.
For N-dimensional arrays, it is a sum product over the 
last axis of a and the second-last axis of b.

a = np.array([[1,2],[3,4]]) 

b = np.array([[11,12],[13,14]]) 

np.dot(a,b)

It will produce the following output −

[[37, 40] 
 [85,92]] 
 
Note that the dot product is calculated as −

[[1x11+2x13, 1x12+2x14],[3x11+4x13, 3x12+4x14]]
