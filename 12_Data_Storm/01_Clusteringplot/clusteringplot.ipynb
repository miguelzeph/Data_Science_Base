```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = {
    'mês':['jan','jun','dez','jan','jun','dez','jan','jun','dez','jan','jun','dez','jan','jun','dez','jan','jun','dez'],
    'ano':[2001,2001,2001,2002,2002,2002,2003,2003,2003,2004,2004,2004,2005,2005,2005,2006,2006,2006],
    'lucro':[2700,2100,3150,1000,1500,2000,3000,1800,1100,1500,1700,2000,1500,1800,2500,3000,1000,3400],
}
```

```python
df = pd.DataFrame(dados)
df.head(2)
```

```python
#Transformar em uma matriz

df1 = df.pivot_table(index = 'mês', columns = 'ano', values = 'lucro')
df1
```

clustermap
https://seaborn.pydata.org/generated/seaborn.clustermap.html

```python
sns.clustermap(data = df1,annot = True,fmt='d' ,cmap="YlGnBu")
#fmt='.2f'
```

Metrics (scipy)

‘braycurtis’, ‘canberra’, ‘chebyshev’, ‘cityblock’, ‘correlation’, ‘cosine’, ‘dice’, ‘euclidean’, ‘hamming’, ‘jaccard’, ‘jensenshannon’, ‘kulsinski’, ‘mahalanobis’, ‘matching’, ‘minkowski’, ‘rogerstanimoto’, ‘russellrao’, ‘seuclidean’, ‘sokalmichener’, ‘sokalsneath’, ‘sqeuclidean’, ‘yule’.

#### <font color = 'red'> A escolha das métricas é arbitrária (facultativa), ou seja, não há uma razão biológica ou física para escolher uma ou outra... escolha a que mais lhe ajuda a analisar seus dados!</font>

```python
sns.clustermap(data = df1,annot = True,fmt='d' ,cmap="YlGnBu",metric="braycurtis") #GOSTEI DESTE
#fmt='.2f'
plt.savefig('./graf/cluster1.png')
```

```python
sns.clustermap(data = df1,annot = True,fmt='d' ,cmap="YlGnBu",metric="braycurtis",row_cluster=False,)
#fmt='.2f'
plt.savefig('./graf/cluster2.png')
```

```python
sns.clustermap(data = df1,annot = True,fmt='d' ,cmap="YlGnBu",metric="braycurtis",col_cluster=False,)
#fmt='.2f'
plt.savefig('./graf/cluster3.png')
```

```python
sns.clustermap(data = df1,annot = True,fmt='d' ,cmap="YlGnBu",metric="braycurtis",col_cluster = False,row_cluster=False,)


# Se não aplicar as funções Métricas na Col ou na Row os clusters.. Não faz nada,
# é como se estivesse plotando um simples mapa de calor.
```
