# Pacientes confirmados com Covid-19

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./time_series_covid_19_confirmed.csv')
```

```python
df.head()
```

```python
#Trocar os nomes por mais fáceis
df = df.rename(columns= {"Country/Region" : "Country", "Province/State": "Province"})
```

```python
#Vamos retirar Lat and Long
df = df.drop(['Lat','Long'],axis=1)
```

```python
# Existem países repetidos, devido suas províncias
```

```python
# Vamos classificar tudo no mesmo grupo... e somar os que forem iguais
df_new= df.groupby('Country').sum()
```

```python
df_new.head()
```

```python
# Transpose!
df_new = df_new.transpose()
```

```python
df_new.head()
```

```python
df_new[['Brazil','China']].head()
```

```python
def data(x):
    
    data = str(x.index).split('/')
    escrever = data[1]+'/'+data[0]+"/"+data[2]
    return escrever
```

```python
b = df_new[['Brazil','China','Italy','United Kingdom','Germany','Spain']]
```

```python
b.plot()
```

```python
plt.figure(figsize = (10,6))
plt.plot(
    df_new.index,
    df_new['Brazil'],
    color='black',
    linestyle='-',
    marker = 'o',
    markerfacecolor = 'blue',
    markersize = 7,
    label = 'Brazil'
)
plt.plot(
    df_new.index,
    df_new['Italy'],
    color='black',
    linestyle='-',
    marker = 'o',
    markerfacecolor = 'white',
    markersize = 7,
    label = 'Italy'
)
plt.plot(
    df_new.index,
    df_new['Spain'],
    color='black',
    linestyle='-',
    marker = 'o',
    markerfacecolor = 'orange',
    markersize = 7,
    label = 'Spain'
)
plt.plot(
    df_new.index,
    df_new['Germany'],
    color='black',
    linestyle='-',
    marker = 'o',
    markerfacecolor = 'green',
    markersize = 7,
    label = 'Germany'
)
plt.plot(
    df_new.index,
    df_new['China'],
    color='black',
    linestyle='-',
    marker = 'o',
    markerfacecolor = 'red',
    markersize = 7,
    label = 'China'
)
plt.xticks(np.arange(0,len(df_new),int(len(df_new)/5))) #IMPORTANTE
plt.yticks([0,50000,100000,150000,200000])
plt.title('Pacientes confirmados com Covid-19 (22/01/20 - 22/04/20)',fontsize = 15)
plt.legend()
plt.grid()
plt.savefig('./confirmado.png');
```

```python
cor = ['purple','red','violet','orange']
nome = ['Germany','Brazil','China','Spain']

def plotar(pais,comparacao,c):
    k = 0
    for nome,cor in zip(pais,c):
        b.plot(x = f'{nome}', y = f'{comparacao}',kind='scatter',color = f'{cor}')
        plt.title(f"Casos Confirmados com Covid-19: {comparacao} x {nome}")
        plt.grid()
        plt.savefig('./comp_'+str(k)+'_.png')
        #plt.close()
        k+=1
plotar(nome,'Italy',cor);       
```
