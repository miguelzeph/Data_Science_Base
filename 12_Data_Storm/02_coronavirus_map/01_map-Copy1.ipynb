# Part 1: Scatter Plots on Maps

```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv('./time_series_covid_19_confirmed.csv')
```

```python
df.head()
```

```python
# Renomear algumas colunas
df = df.rename(columns = {
    "Country/Region":'Country',
    "Province/State":'Province'})
```

```python
df.head()
```

```python
# Criei uma coluna nova com os dados de HOJE
df['text'] = df['Country'] + " " + df["4/22/20"].astype(str)
```

```python
df.head()
```

```python
fig = go.Figure(data = go.Scattergeo(
    lon = df["Long"],
    lat = df["Lat"],
    text = df["text"],
    mode = "markers",
    marker = dict(
        size = 12,
        opacity = 0.8,
        reversescale = True,
        autocolorscale = True,
        symbol = 'square',
        line = dict(
            width = 1,
            color = 'rgba(102, 102, 102)'
        ),
        cmin = 0,
        color = df['4/22/20'],
        cmax = df['4/22/20'].max(),
        colorbar_title = "COVID 19 Reported Cases"
    )
))

fig.update_layout(
    title = "COVID19 Confirmed Cases Around the World",
    geo = dict(
        scope = "world",
        showland = True,
    )
)

fig.write_html('first_figure.html', auto_open=True)
```

# Part 2: Choropleth Map

```python
# Creat a unique country list
```

```python
#importar dados
df = pd.read_csv("./time_series_covid_19_confirmed.csv")

# trocar nome da coluna
df = df.rename(columns= {"Country/Region" : "Country", "Province/State": "Province"})

# Soma todos os nomes de paÃ­ses iguais
total_list = df.groupby('Country')['4/13/20'].sum().tolist()

# Adiciona os nomes dos paÃ­ses
country_list = df["Country"].tolist()

# Remove os paÃ­ses com nome repetido
country_set = set(country_list)

# Adiciona a lista
country_list = list(country_set)

# Coloca em ordem alfabÃ©tica
country_list.sort()

new_df = pd.DataFrame(list(zip(country_list, total_list)), 
               columns =['Country', 'Total_Cases'])
new_df.head()
```

```python
colors = ["#F9F9F5", "#FAFAE6", "#FCFCCB", "#FCFCAE",  "#FCF1AE", "#FCEA7D", "#FCD97D",
          "#FCCE7D", "#FCC07D", "#FEB562", "#F9A648",  "#F98E48", "#FD8739", "#FE7519",
          "#FE5E19", "#FA520A", "#FA2B0A", "#9B1803",  "#861604", "#651104", "#570303",]


fig = go.Figure(data=go.Choropleth(
    locationmode = "country names",
    locations = new_df['Country'],
    z = new_df['Total_Cases'],
    text = new_df['Total_Cases'],
    #colorscale = colors,
    autocolorscale=False,
    reversescale=False,
    colorbar_title = 'Reported Covid-19 Cases',
))

fig.update_layout(
    title_text='Reported Covid-19 Cases',
    geo=dict(
        showcoastlines=True,
    ),
)

fig.write_html('second_figure.html', auto_open=True)
```
