### Projeto 2

Você tem uma máquina de sorvete e deseja criar um modelo para prever as vendas diárias baseadas na temperatura.

X -> temperatura

Y -> vendas diárias em dólares

```python
# 1-) IMPORTAR BIBLIOTECAS
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
import tensorflow as tf
```

```python
# 2-) VISUALIZAR DADOS
```

```python
sales_df = pd.read_csv('./dados/vendas.csv')
```

```python
sales_df.head()
```

```python
sales_df.tail()
```

```python
sales_df.info()
```

```python
sns.scatterplot(sales_df['Temperature'], sales_df['Revenue'])
```

```python
# 3-) ORGANIZAR DADOS PARA TREINAMENTO
```

```python
X_train = sales_df['Temperature']
y_train = sales_df['Revenue']
```

```python
X_train.shape
```

```python
# 4-) CONSTRUIR SUAS CAMADAS
```

```python
model = tf.keras.Sequential()

# Camada 1
model.add(
    tf.keras.layers.Dense(#Dense= 1 N conecta all
        units = 6, # Saida (Ela se torna a oculta, pois ela é a de entrada da pŕoxima... já que foi criada abaixo)
        input_shape = [1], #Entrada
    )
)
# Esse problema é mais complexo do que o anterior,
#precisaremos de mais camadas ocultas

# Camada 2
model.add(
    tf.keras.layers.Dense(#Dense= 1 N conecta all
        units =1, # Saida
        input_shape = [6]
    )
)

```

<img src = "img/camada.png" style="max-width:80%; width: 80%;">

```python
model.summary()
```

```python
# 5-) TREINAR O MODELO
```

```python
model.compile(
    optimizer= tf.keras.optimizers.Adam(0.05),
    loss= "mean_squared_error"
)
```

```python
epoches_hist = model.fit(
    X_train,
    y_train,
    epochs= 500
)
```

```python
epoches_hist.history.keys()
```

```python
# 6-) VISUALIZAR O RESULTADO DO TREINO
```

```python
plt.plot(epoches_hist.history['loss'][10:])#Tirei os 10 primeiros
plt.title('Modelo de Perda em Treinamento')
plt.xlabel('epocas')
plt.ylabel('Training Loss')
plt.legend('Training Loss')
```

```python
# Visualizar os pesos
pesos = model.get_weights()
pesos
```

```python
peso_df = pd.DataFrame(pesos[0])
peso_df#6 pesos
```

```python
temp = -1 # -1°C
vendas = model.predict([temp]) #temos 23 vendas
vendas
```

```python
model.predict(X_train)
```

```python
#Destar os dados 
X_train[0],model.predict(X_train)[:,0][0]
```

```python
plt.scatter(X_train,y_train,color='blue',label='Experimental')
plt.plot(X_train,model.predict(X_train)[:,0],'-r',label = 'previsto')
plt.xlabel('Temperatura')
plt.ylabel('Vendas')
plt.legend()
```

### Parte 2 - Treinar com Scikit-learn também

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train.shape #Formato de Lista
```
