### Projeto - Prever a formula da relação entre graus celsius e fahrenheit

Iremos fazer ela encontrar a Fórmula de Fahrenheits (y) em função de graus Celsius (x)

x - Variável indepêndente

y - Variável dependente

Técnica de regressão

Iremos fazer uma engenharia reversa... temos os dados respectivos através da formula, mas iremos treinar nosso NN para aprender

```python
import tensorflow as tf
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2-) Importar Banco de Dados
```

```python
temperatura_df = pd.read_csv('./dados/celsius.csv')
temperatura_df.reset_index(drop = True, inplace = True)
```

```python
# 3-) Configuração para treinamento
```

```python
X_train = temperatura_df['Celsius']
y_train = temperatura_df['Fahrenheit']
```

```python
model = tf.keras.Sequential()#sequencia de camadas

# N = neurônio
model.add(
    tf.keras.layers.Dense(#Dense=> 1 N é ligado a todos N da próxima camada
        units = 1, #1 N de saída °F
        input_shape = [1],# 1 N de entrada °C
        )
) 
model.summary()
```

<img src = "img/camada.png" style="max-width:50%; width: 30%;">

```python
model.compile(
    optimizer= tf.keras.optimizers.Adam(0.3),#Taxa de aprendizagem
    loss = 'mean_squared_error'
)
```

```python
epochs_hist = model.fit(
    X_train,
    y_train,
    epochs= 1800,#Número de épocas
)
```

```python
# 5-) Avaliação do Modelo
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'], label = 'loss')
plt.title('Model Loss Progress During Training')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel("Training_Loss")
plt.grid()
```

Não é necessário treinar por mais de 250 épocas esse modelo, pois ele estabiliza a partir de 250.
