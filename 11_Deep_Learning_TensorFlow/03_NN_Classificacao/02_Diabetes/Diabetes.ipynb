### Organizar dados

```python
from sklearn.model_selection import train_test_split
```

```python
X = dados.drop('Outcome',axis=1)
y = dados['Outcome']
```

```python
X.shape,y.shape
```

```python
y = np.array(y)
y.reshape(-1,1)[:4] #y virou matriz em vez de lista
```

###### Ajustando Escalas eixo X (É bom para melhorar o processamento)

## DICA: melhora muito o aprendizado da máquina! teste fazer sem ajustar!

```python
from sklearn.preprocessing import StandardScaler
#Standard Scale - ele padroniza com desvio 
#padrão (pode assumir valores negativos)
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

#### ----- fim do ajuste de escala ------

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2,    
)
```

```python
X_train.shape, y_train.shape,X_test.shape,y_test.shape
```

### Criar Modelo

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(
    units=400,
    activation = 'relu',
    input_shape = (X_train.shape[1],),
    )
)
model.add(tf.keras.layers.Dropout(0.2)) # Ele pega 20% dos dados da camada de cima e pula eles
#Dropout evita que o modelo se adpte demais aos dados de Treinamento!

model.add(tf.keras.layers.Dense(
    units=400,
    activation = 'relu',
    )
)
model.add(tf.keras.layers.Dropout(0.2)) # Ele pega 20% dos dados da camada de cima e pula eles
#Dropout evita que o modelo se adpte demais aos dados de Treinamento!


model.add(tf.keras.layers.Dense(
    units = 1,
    activation = 'sigmoid',
    )
)

model.summary()
```

```python
model.compile(loss = 'binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs= 200,
)
```

```python
#Aqui estão armazenados os valores de erro e acerto
history.history.keys()
```

```python
plt.figure(figsize = (15,6))
plt.subplot(1,2,1)
plt.plot(history.history['loss'],'-or')
plt.xlabel('epochs')
plt.ylabel('loss')

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], '-ob')
plt.xlabel('epochs')
plt.ylabel('accuracy')
```

### Observações do Gráfico acima

Quanto maior a camada, maior será a variação dos dados, porém maior será o acerto.
Quanto menos camadas, menor será a variação dos dados, porém menor será o acerto.

Quanto mais camadas ocultas ter, mais demorado é o aprendizado, ou seja, demora para overlapar um valor médio, porém maior será o acerto a longo prazo.

### Predictions

```python
pred = model.predict(X_train)

pred = (pred > 0.5)

pred
#maior que 0.5 é da classe positiva e menor da classe negativa
```

```python
df_pred = pd.DataFrame()
df_pred['pred'] = pred[:,0]
df_pred['y_train'] = y_train
```

### Matriz de Confusão (vamores verificar os acertos)

```python
from sklearn.metrics import confusion_matrix
```

```python
y_train_pred = model.predict(X_train)
y_train_pred = (y_train_pred >0.5)


cm = confusion_matrix(y_train,y_train_pred)
sns.heatmap(cm,annot=True)
```

```python
cm
```

### Analise de Classificação

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train_pred, y_train))
```

### Vamos a Base de Teste agora

```python
y_test_pred = model.predict(X_test)
y_test_pred = (y_test_pred >0.5)


cm = confusion_matrix(y_test,y_test_pred)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test_pred, y_test))
```
