```python
def plot_comparison(x, title):
    fig, ax = plt.subplots(3, 1, sharex=True)
    sns.distplot(x, ax=ax[0])
    ax[0].set_title('Histogram + KDE')
    sns.boxplot(x, ax=ax[1])
    ax[1].set_title('Boxplot')
    sns.violinplot(x, ax=ax[2])
    ax[2].set_title('Violin plot')
    fig.suptitle(title, fontsize=16)
    plt.show()
```

```python
np.random.seed(42)# Faz repetir os números
sample_gaussian = np.random.normal(size=10000) # Vai gerar números repetidos


plot_comparison(sample_gaussian, 'Standard Normal Distribution')
```

### Gráfico Violinio + Plot das Medianas

```python
# Pegar valores das medianas por dia
df = pd.DataFrame()
df['day'] = tips['day'].unique()
```

```python
tips['day'].unique()# Está fora da ordem... 
```

```python
y = []
#for day in tips['day'].unique():
for day in ['Thur','Fri','Sat','Sun']: # único jeito que consegui organizar
    x = tips[tips['day'] == day].median()
    y.append(x['total_bill'])
print(y)
df['total_bill'] = y
```

```python

fig, ax = plt.subplots(figsize=(8,6))
sns.violinplot(
    x='day',
    y = 'total_bill',
    data = tips,
    scale= 'area'
)

#df.plot(ax=ax)
ax.plot(np.array(df['day']),np.array(df['total_bill']),
        linestyle='dashed',
        color = 'black',
        linewidth=3,
        label = 'Mediana'
       )
ax.legend()
ax.set_xlim(-1,4);
ax.set_ylim(-5,60)
```
