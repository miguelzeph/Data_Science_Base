# Logistic Regression - parte 2

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_data = pd.read_csv('./dados_panda/housing.csv')

# DataFrame.sample = é diferente do head
#retorna índices aleatórios
housing_data.sample(5)
```

```python
housing_data = housing_data.dropna() #Remove os NaN
```

```python
housing_data.shape
```

```python
housing_data.loc[housing_data['median_house_value'] == 500001].head()
```

```python
# Access a group of rows and columns by label(s) or a boolean array.
housing_data.loc[housing_data['median_house_value'] == 500001].count()
```

```python
housing_data = housing_data.drop(housing_data.loc[housing_data['median_house_value']== 500001].index)
```

```python
housing_data.head()
```

```python
housing_data['ocean_proximity'].unique()
```

```python
# Convert categorical variable into dummy/indicator variables
housing_data = pd.get_dummies(housing_data, columns = ['ocean_proximity'])
```

```python
housing_data.sample(10)
```

```python
median = housing_data['median_house_value'].median()
median
```

```python
housing_data['above_median'] = (housing_data['median_house_value'] - median) > 0
```

```python
housing_data.sample(5)
```

```python
X =(housing_data.drop(['median_house_value','above_median'],axis=1))
#y = pd.DataFrame(housing_data['above_median'])
#ou
y = housing_data['above_median']
```

```python
X.columns
```

```python
#y.columns #Tem que fazer y virar DataFrame
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression(
    solver = 'liblinear',
).fit(X_train,y_train)
```

```python
logistic_model.score(X_train,y_train)
```

```python
logistic_model.score(X_test,y_test)
```

```python
y_pred = logistic_model.predict(X_test)
```

```python
df_pred_actual = pd.DataFrame({'predicted': y_pred, 'actual': y_test})
df_pred_actual.head(5)
```
